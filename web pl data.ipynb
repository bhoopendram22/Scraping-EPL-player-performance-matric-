```python
import pandas as pd
import requests
from bs4 import BeautifulSoup

```

```python
url="https://www.premierleague.com/stats/top/players/goals"
r=requests.get(url)
print(r)
```

```python
soup=BeautifulSoup(r.text,"lxml")
#print(soup)
```

```python
name=soup.find_all("a",class_="playerName")
#print(name.text)
ply_name=[]
for i in name:
    name=i.text.strip()
    ply_name.append(name)
print(ply_name)
```

```python

clb=soup.find_all("a",class_="stats-table__cell-icon-align")
#print(name.text)
club_name=[]
for i in clb:
    clb=i.text.strip()
    club_name.append(clb)
print(club_name)
```

```python


nation=soup.find_all("div",class_="stats-table__cell-icon-align")
#print(name.text)
na_name=[]
for i in nation:
    nation=i.text.strip()
    na_name.append(nation)
print(na_name)
```

```python
ast=soup.find_all("td",class_="stats-table__main-stat")
#print(name.text)
ast_name=[]
for i in ast:
    ast=i.text.strip()
    ast_name.append(ast)
print(ast_name)
```

```python

assist=pd.DataFrame({"palyer name":ply_name,"club":club_name,"nationality":na_name,"assists":ast_name})
assist
```

```python
assist.to_csv('new.pl_most.assists.csv')
```
